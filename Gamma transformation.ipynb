```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s_img = cv2.imread("/home/ahmed/Desktop/wordpress/spine.png")
```

```python
img = cv2.cvtColor(s_img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
gamma = 1
img_2 = np.power(img,gamma)
plt.imshow(img_2, cmap='gray')
```
