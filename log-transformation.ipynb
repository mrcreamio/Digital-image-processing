```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s_img = cv2.imread("/home/ahmed/Desktop/wordpress/man.jpeg")
```

```python
img = cv2.cvtColor(s_img, cv2.COLOR_BGR2RGB)
plt.imshow(s_img)
```

```python
img_1 = np.uint8(np.log1p(s_img))
thresh = 1
img_2 = cv2.threshold(img_1,thresh,255,cv2.THRESH_BINARY)[1]
plt.imshow(img_2)
```

```python
normalized_image = cv2.normalize(img_1, None, 0, 255, cv2.NORM_MINMAX, dtype = cv2.CV_8U)
plt.imshow(normalized_image)
```
