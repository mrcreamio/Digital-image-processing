```python
def flat(list1):
    for i in list1:
        for j in i:
            if (j != 1 and j != 0):
                return False
    return True
def match(img, ker):
    if img == ker:
        return True
    else:
        return False
    
def oneD(nested):
    flat = [x for sublist in nested for x in sublist]
    return flat
def sub(ziped):
    ret = []
    for i in ziped:
        ret.append(i[0]-i[1])
    return ret
def erosion(image,kernel,a,b):
    ker = kernal.tolist()
    imgL = image.tolist()
    img_x, img_y = image.shape
    kernl_x, kernl_y = kernel.shape
    result = copy.deepcopy(image)
    for i in range(1,(img_x)-int(len(kernel)/2)):
        for j in range(1,(img_y)-int(len(kernel)/2)):
            array = []
            for s in range(-a,a+1):
                for t in range(-b,b+1):
                    array.append(image[i+s,j+t])
            if(flat(kernel)):
                if(flat(imgL)):
#                     print("kernal is flat and image is binary")
                    if(match(array,ker)): #if any one element matched it will write 1 on its place
                        result[i,j] = 1
                    else:
                        result[i,j] = 0
                else:
#                     print("kernal is flat and image is grey scale")
                    mini = min(array) # maximum number from the kernal will be placed on the interest point
                    result[i,j] = mini
            else:
#                 print("kernal is non flat")
                ziped =  zip(array,oneD(ker))#subtracting kenal and window elements of image
                mini = min(sub(ziped))
                result[i,j] = mini
                

                
    return result
def histogram(img):
    plt.hist(img.ravel(),256,[0,256]) 
    plt.plot
    plt.show()

    
```
