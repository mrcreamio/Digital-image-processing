```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import copy
import math
from tkinter import *
```

```python
def flat(list1):
    for i in list1:
        for j in i:
            if (j != 1 and j != 0):
                return False
    return True
def match(img, ker):
    counter = 0
    for i in range(len(ker)):
        for j in range(len(ker)):
            if img[counter] == ker[i][j]:
                counter+=1
                return True
            else:
                counter+=1
    else:
        return False
    
def oneD(nested):
    flat = [x for sublist in nested for x in sublist]
    return flat

def dilation(image,kernel,a,b):
    ker = kernal.tolist()
    imgL = image.tolist()
    img_x, img_y = image.shape
    kernl_x, kernl_y = kernel.shape
    result = copy.deepcopy(image)
    for i in range(1,(img_x-1)-int(len(kernel)/2)):
        for j in range(1,(img_y-1)-int(len(kernel)/2)):
            array = []
            for s in range(-a,a+1):
                for t in range(-b,b+1):
                    array.append(image[i+s,j+t])
            if(flat(kernel)):
                if(flat(imgL)):
#                     print("kernal is flat and image is binary")
                    if(match(array,ker)): #if any one element matched it will write 1 on its place
                        result[i,j] = 1
                    else:
                        result[i,j] = 0
                else:
#                     print("kernal is flat and image is grey scale")
                    maxi = max(array) # maximum number from the kernal will be placed on the interest point
                    result[i,j] = maxi
            else:
#                 print("kernal is non flat")
                res_list = [sum(i) for i in zip(array,oneD(ker))]  #adding kenal and window elements of image
                maxi = max(res_list)
                result[i,j] = maxi
            
    return result
```

```python

def matchE(img, ker):
    if img == ker:
        return True
    else:
        return False
    
def sub(ziped):
    ret = []
    for i in ziped:
        ret.append(i[0]-i[1])
    return ret
def erosion(image,kernel,a,b):
    print(kernel)
    ker = kernal.tolist()
    imgL = image.tolist()
    img_x, img_y = image.shape
    print(image.shape)
    kernl_x, kernl_y = kernel.shape
    print(int(len(kernal)/2))
    result = copy.deepcopy(image)
    for i in range(1,(img_x-1)-int(len(kernel)/2)):
        for j in range(1,(img_y-1)-int(len(kernel)/2)):
            array = []
            for s in range(-a,a+1):
                for t in range(-b,b+1):
                    array.append(image[i+s,j+t])
            if(flat(kernel)):
                if(flat(imgL)):
#                     print("kernal is flat and image is binary")
                    if(matchE(array,ker)): #if any one element matched it will write 1 on its place
                        result[i,j] = 1
                    else:
                        result[i,j] = 0
                else:
#                     print("kernal is flat and image is grey scale")
                    mini = min(array) # maximum number from the kernal will be placed on the interest point
                    result[i,j] = mini
            else:
#               print("kernal is non flat")
                ziped =  zip(array,oneD(ker))#subtracting kenal and window elements of image
                mini = min(sub(ziped))
                result[i,j] = mini
                

                
    return result
```
